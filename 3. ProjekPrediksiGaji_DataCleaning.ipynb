```python
#assigning minimum and maximum value of years of experiences
min_years = []
max_years = []
y = []

for x in df_total['Yrs_of_Exp']:
  y.append(str(x).split())

for x in y:
  if (x[0].isdigit()):
    if len(x)==2:
      min_years.append(0)
      max_years.append(0)
    else:
      min_years.append(x[0])
      max_years.append(x[2])
  elif (x[0] == 'Kurang')|(x[0] == 'Setidaknya'):
    min_years.append(0)
    max_years.append(1)
  elif x[0] == 'Lebih':
    min_years.append(10)
    max_years.append(20)
  elif x[0] == 'nan':
    min_years.append('')
    max_years.append('')
```

```python
#cleaning salary column
df_total['Salary'].value_counts()
```

```python
#assigning minimum and maximum salary
df_total['Salary'] = df_total['Salary'].replace('Login to view estimated salary by Glints.',np.nan)
df_total['Salary'] = df_total['Salary'].replace('Gaji Kompetitif',np.nan)
df_total = df_total.dropna(axis=0)
df_total.reset_index(drop=True, inplace=True)

y = []
min_salary = []
max_salary = []
cc = CurrencyConverter()

for salary in df_total['Salary']:
  y.append(str(salary).strip().replace('.','').split())

for x in y:
   if x[0] == 'IDR':
     min_salary.append(x[1].strip(''))
     max_salary.append(x[-1])
   elif x[0] == '0':
     min_salary.append(0)
     max_salary.append(0)
   elif x[0] == 'USD':
     min_salary.append(cc.convert(x[1],'USD','IDR'))
     max_salary.append(cc.convert(x[-1],'USD','IDR'))
   elif x[0] == 'SGD':
     min_salary.append(cc.convert(x[1],'SGD','IDR'))
     max_salary.append(cc.convert(x[-1],'SGD','IDR'))
   elif x[0] == 'AUD':
     min_salary.append(cc.convert(x[1],'AUD','IDR'))
     max_salary.append(cc.convert(x[-1],'AUD','IDR'))
   elif x[0] == 'HKD':
     min_salary.append(cc.convert(x[1],'HKD','IDR'))
     max_salary.append(cc.convert(x[-1],'HKD','IDR'))
```

```python
#join all dataframes together
df_cleaned_data = pd.concat([df_total['Job'],df_total['Province'],pd.DataFrame(min_years),pd.DataFrame(max_years),pd.DataFrame(min_salary),pd.DataFrame(max_salary)],axis=1)
df_cleaned_data.columns=['Job','Province','Min_Years','Max_Years','Min_Salary','Max_Salary']
df_cleaned_data
```
