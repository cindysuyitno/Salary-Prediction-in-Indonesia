```python
#library imports
from bs4 import BeautifulSoup
import requests
import lxml
import pandas as pd
import numpy as np
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
import os
from random import randint
from time import sleep
```

```python
#these bunch of codes are important to establish webdriver.Chrome, which is a python tool for web scraping
!cp /usr/lib/chromium-browser/chromedriver /usr/bin
import sys
sys.path.insert(0,'/usr/lib/chromium-browser/chromedriver')

chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
wd = webdriver.Chrome('chromedriver',chrome_options=chrome_options)
```

```python
#this is the data href i scrapped on 6/9/2022, on karir.com (which has 132 pages)
karir_site = []
for a in range(132):
  karir_site.append(f'https://www.karir.com/search?q=&sort_order=urgent_job&job_function_ids=&industry_ids=&degree_ids=&major_ids=&location_ids=&location_id=&location=&salary_lower=0&salary_upper=100000000&page={a+1}&grid=list')
```

```python
#there are 6 items i want to scrap from this website: job name, company, department, city, salary, and years of experiences
#these are the class name of those items (can be retrieved from click right -> Inspect to the text in the website)
#job_name = tdd-function-name --semi-bold --inherit;
#company = tdd-company-name h8 --semi-bold;
#department = tdd-company;
#city = tdd-location;
#salary = tdd-salary;
#yrs_of_experience = tdd-experience.

#this code should work. but when i run it there are some problems where i got data that i don't want, 
#hence i try to do each item one by one in the following code.
my_table_data2 = []
for i in range(132):
  url = karir_site[i]
  wd.get(url)
  soup = BeautifulSoup(wd.page_source, 'html.parser')
  my_table = soup.find_all(class_=['tdd-function-name --semi-bold --inherit','tdd-company-name h8 --semi-bold','tdd-company','tdd-location','tdd-salary','tdd-experience'])
  for data in my_table:
      my_table_data2.append(data.get_text())
```

```python
my_table_data2_1 = []
my_table_data2_2 = []
my_table_data2_3 = []
my_table_data2_4 = []
my_table_data2_5 = []
my_table_data2_6 = []

for i in range(132):
  url = karir_site[i]
  wd.get(url)
  soup = BeautifulSoup(wd.page_source, 'html.parser')
  my_table1 = soup.find_all(class_=['tdd-function-name --semi-bold --inherit'])
  for data in my_table1:
      my_table_data2_1.append(data.get_text())
 
  my_table2 = soup.find_all(class_=['tdd-company-name h8 --semi-bold'])
  for data in my_table2:
      my_table_data2_2.append(data.get_text())

  my_table3 = soup.find_all(class_=['tdd-company'])
  for data in my_table3:
      my_table_data2_3.append(data.get_text())

  my_table4 = soup.find_all(class_=['tdd-location'])
  for data in my_table4:
      my_table_data2_4.append(data.get_text())
  
  my_table5 = soup.find_all(class_=['tdd-salary'])
  for data in my_table5:
      my_table_data2_5.append(data.get_text())
  
  my_table6 = soup.find_all(class_=['tdd-experience'])
  for data in my_table6:
      my_table_data2_6.append(data.get_text())
```

```python
#assigning each item to dataframe with suitable column names
df = pd.DataFrame()
df['Job_Title']=my_table_data2_1
df['Company']=my_table_data2_2
df['Dept']=my_table_data2_3
df['City']=my_table_data2_4
df['Salary']=my_table_data2_5
df['Yrs_of_Exp']=my_table_data2_6
```

```python
#redoing the dataframe input
df = pd.DataFrame()
df['Job_Title']=my_table_data2_1
df['Company']=my_table_data2_2
df['Dept']=my_table_data2_3
df['City']=my_table_data2_4new
df['Salary']=my_table_data2_5
df['Yrs_of_Exp']=my_table_data2_6
```
